# **capillary Stable**

```python
import pandas as pd
df=pd.read_excel('/content/drive/MyDrive/Bearing/cappilary stable.xlsx')
df
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming your DataFrame is named 'df'
# Replace df with your actual DataFrame name

# Scatter plot between columns 'a' and 'b'
plt.figure(figsize=(8, 6))
plt.scatter(df['a'], df['b'], color='blue', alpha=0.7)
plt.title('Scatter plot of a vs b')
plt.xlabel('Column a')
plt.ylabel('Column b')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming your DataFrame is named 'df'
# Replace df with your actual DataFrame name

# Extracting 'a' and 'b' columns in sets of 30 readings
num_readings = 30
total_rows = len(df)
num_sets = total_rows // num_readings

# Plotting in sets of 30 readings
plt.figure(figsize=(10, 6))
for i in range(num_sets):
    start_idx = i * num_readings
    end_idx = start_idx + num_readings
    subset = df.iloc[start_idx:end_idx]
    plt.plot(subset['a'], subset['b'], label=f'Set {i+1}')

plt.title('Plot of a and b in sets of 30 readings')
plt.xlabel('Column a')
plt.ylabel('Column b')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming your DataFrame is named 'df'
# Replace df with your actual DataFrame name

num_readings = 30
total_rows = len(df)
num_sets = total_rows // num_readings

fig, axes = plt.subplots(num_sets, figsize=(8, 5*num_sets))

for i, ax in enumerate(axes.flatten()):
    start_idx = i * num_readings
    end_idx = start_idx + num_readings
    subset = df.iloc[start_idx:end_idx]

    del_val = subset['del'].values[0]
    w_val = subset['w'].values[0]
    Mj_val = subset['Mj'].values[0]
    x_val = subset['x'].values[0]
    y_val = subset['y'].values[0]

    ax.plot(subset['a'], subset['b'])
    ax.set_title(f'Set {i+1}: del={del_val}, w={w_val}, Mj={Mj_val}, x={x_val}, y={y_val}')
    ax.set_xlabel('Column a')
    ax.set_ylabel('Column b')
    ax.grid(True)

plt.tight_layout()
plt.show()
```
